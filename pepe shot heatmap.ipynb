```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('pepe_shotdata.csv')
data.head()
```

```python
df = data.drop(['Unnamed: 0'], axis=1)
df = data[(data['Season']==2020)]
df
```

```python
df['x'] = df['x']*100*1.2
df['y'] = df['y']*100*.8
df.head()
```

```python
from mplsoccer.pitch import Pitch
```

```python
pitch = Pitch(pitch_color='black', line_color='white', stripe=False, figsize=(13, 8))
fig, ax = pitch.draw()

#Draw the pitch on the ax figure as well as invert the axis for this specific pitch
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

#Create the heatmap
kde = sns.kdeplot(
        df['x'],
        df['y'],
        shade = True,
        shade_lowest=False,
        alpha=.5,
        n_levels=10,
        cmap = 'magma'
)
plt.xlim(0,120)
plt.ylim(0,80)
plt.title("Pepe Shot HeatMap",color='Black',size=20)
```

```python
pepe_shottype = df[['xG','ShotType']]
pepe_shottype = pepe_shottype.groupby(['ShotType']).sum()
pepe_shottype = pepe_shottype.sort_values('xG', ascending=False).reset_index()

pepe_shottype
```

```python
#To see which body type Pepe mostly use in 2020 with ShotType data

sns.barplot(x='ShotType',y='xG',data=pepe_shottype,color="blue")
plt.xlabel('ShotType', size=14)
plt.ylabel('xG', size=14)
plt.title("Pepe Shot Type 2020", size=20)
plt.xticks(size=11)
plt.show()
```
